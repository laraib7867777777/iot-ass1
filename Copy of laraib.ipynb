```python
#Q9

n=int(input("Enter any number for factorial: "))
factorial=1
if n<0:
  print("sorry factorial doesnt exist for negative number !")
elif n==0:
  print(" the factorial of 0 is 1 ")
else:
  for i in range(1,n+1): 
     factorial= factorial*i
print(factorial)
  
```
